# Aula 2: SQL - Exercícios

```python
import yaml
import sqlalchemy
import pandas as pd
```

```python
# Load SQL credentials
with open('credentials.yml', 'r') as file:
    credentials = yaml.load(file, Loader=yaml.FullLoader)
```

```python
engine = sqlalchemy.create_engine(f'mysql+mysqlconnector://{credentials["user"]}:{credentials["password"]}@{credentials["host"]}:{credentials["port"]}/{credentials["database"]}') # connect to server
```

```python
query = '''
SHOW COLUMNS FROM AtrasoCartaoCredito
'''
pd.read_sql_query(query, engine)
```

#### Selecione as colunas `Target` e `Sexo`

```python
query = '''
SELECT Target, Sexo
FROM AtrasoCartaoCredito
'''
pd.read_sql_query(query, engine)
```

#### Selecione as colunas `Target` e `Sexo` com pessoas de idade acima de 30 anos

```python
query = '''
SELECT Target, Sexo
FROM AtrasoCartaoCredito
WHERE Idade > 30
'''
pd.read_sql_query(query, engine)
```

#### Selecione as colunas `Target` e `Sexo` com pessoas de idade entre 30 e 50 anos

```python
query = '''
SELECT Target, Sexo
FROM AtrasoCartaoCredito
WHERE Idade between 30 and 50
'''
pd.read_sql_query(query, engine)
```

#### Selecione pessoas da região nordeste

```python
query = '''
SELECT *
FROM AtrasoCartaoCredito
WHERE LOWER(RegiaodoPais) like '%nordeste%'
'''
pd.read_sql_query(query, engine)
```

#### Selecione pessoas das regiões nordeste e sudeste

```python
query = '''
SELECT *
FROM AtrasoCartaoCredito
WHERE LOWER(RegiaodoPais) like '%nordeste%' OR LOWER(RegiaodoPais) LIKE '%sudeste%'
'''
pd.read_sql_query(query, engine)['RegiaodoPais'].value_counts()
```

#### Selecione pessoas de todas regiões exceto sul

```python
query = '''
SELECT *
FROM AtrasoCartaoCredito
WHERE LOWER(RegiaodoPais) not like '%sul%'
'''
pd.read_sql_query(query, engine)['RegiaodoPais'].value_counts()
```

#### Selecione pessoas de MG e SP, e com idade acima de 30 anos

```python
query = '''
SELECT *
FROM AtrasoCartaoCredito
WHERE UPPER(UF) IN ('MG','SP')
AND Idade > 30
'''
pd.read_sql_query(query, engine)
```

#### Selecione pessoas de sexo `homem`, residentes em  RJ e com idade abaixo de 50 anos

```python
query = '''
SELECT *
FROM AtrasoCartaoCredito
WHERE LOWER(Sexo) = 'homem'
AND UPPER(UF) = 'RJ'
AND Idade < 50
'''
pd.read_sql_query(query, engine)
```
