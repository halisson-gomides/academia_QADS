4. Crie uma função que recebe uma string devolve a frequência absoluta de cada palavra da string. Antes da contagem, faça um tratamento prévio do texto colocando todas as palavras em letras minúsculas. A função abaixo gera os dados para resolução da questão (não alterar).

```python
import requests as r
from zipfile import ZipFile
import io
import pandas as pd

def get_data_questao04():
    file_zipped = r.get('https://archive.ics.uci.edu/ml/machine-learning-databases/00331/sentiment%20labelled%20sentences.zip')
    z = ZipFile(io.BytesIO(file_zipped.content))
    z.extractall("/content/")

    return ' '.join(pd.read_csv('/content/sentiment labelled sentences/imdb_labelled.txt',sep='\t',names=['texto','class']).texto.values)

texto_para_a_questao = get_data_questao04()
```

```python
# setando caracteres que quero excluir das palavras
char_to_replace = {',': '', '(': '', ')': ''}

def freq_palavras(texto:str)->dict:
    lista = texto.upper().translate(str.maketrans(char_to_replace)).split()
    dict_count = {x: lista.count(x) for x in lista}
    return dict_count

print(freq_palavras(texto_para_a_questao))

# texto_para_a_questao.upper().translate(str.maketrans(char_to_replace)).split()
# dir(texto_para_a_questao)
```

8. Crie um função que recebe uma lista e retorna o seu desvio padrão.

```python
def func_std_dev(lista:list):
    # calculando a media
    try:
        media = sum(lista) / len(lista)
    except ZeroDivisionError:
        return 0
    
    # calculando o quadrado da distancia entre cada elemento e a média
    sqr_dist = [(x - media)**2 for x in lista]
    
    # somando os valores da etapa anterior, dividir pelo tamanho populacional e calcular a raiz quadrada
    std_dev = (sum(sqr_dist) / len(lista))**0.5
    
    return std_dev


print(func_std_dev(lista_original))
```

9. SDR (Standard Deviation Reduction) é o cálculo de redução do desvio padrão depois da quebra de um dataset utilizando uma regra. Ele consiste na seguinte fórmula.

sdr.png

Onde :

*   SDR(T,X) é redução do desvio padrão original T dado a regra X
*   S(T) é o desvio padrão original
*   S(T,X) é o desvio padrão ponderado utilizando a regra X.


(use desvio populacional)



exemplo.png

Utilizando a imagem acima como exemplo. Em um dataset de vários aluguéis utilizamos uma regra de separação quanto ao valor ser maior do que 16. Assim temos dois subsets. O primeiro tem 5 valores e um desvio padrão de 2,17, já o segundo tem 6 valores e um desvio padrão de 47,43. O desvio padrão ponderado de ambos é 26,86. A diferença entre o desvio padrão original (40,16) e dado a regra (26,86) é 13,30. 

Crie um código que calcule o SDR da lista_original para cada um dos possíveis valores de regra entre 0 e 200. Guarde todos os valores de SDR em uma lista e a passe para a função plot_sdr já definida aqui.

```python
def SDR(lista:list):
    list_sdr = []
    for regra in range(201):
        ds_regra = list(filter(lambda x: x > regra, lista))
        ds_residuo = [x for x in lista if x not in ds_regra]
        S = func_std_dev(lista) # desvio padrão original
        Sregra = func_std_dev(ds_regra) # desvio padrão do dataset que atende a regra
        Sn_regra = func_std_dev(ds_residuo) # desvio padrão do dataset que NÃO atende a regra
        Spond = (len(ds_regra)/ len(lista))*Sregra + (len(ds_residuo)/ len(lista))*Sn_regra # desvio padrão ponderado
        list_sdr.append(S - Spond)
        
    return list_sdr
```

```python
lista_SDR = SDR(lista_original)
```

```python
import matplotlib.pyplot as plt

def plot_sdr(lista_de_sdr):
    f,ax = plt.subplots(1,1,figsize=(16,6))
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)


    ax.set_title('Comportamento de SDR',fontsize=25)
    ax.set_ylabel('SDR',fontsize=15)
    ax.plot(lista_de_sdr)

    plt.show()
```

```python
plot_sdr(lista_SDR)
```

```python
# Defina um dicionário com os o estoque de cada alimento abaixo.
# banana - 10
# maça - 2
# uva - 0
# beterraba - 11
# ovos - 0
# bacon - 5
# pão integral - 12
dict_estoque = {'banana': 10, 'maça': 2, 'uva': 0, 'beterraba': 11, 'ovos': 0, 'bacon': 5, 'pão integral': 12}

#  Defina outro dicionário com os preços de cada alimento abaixo.

# banana - 2.99
# maça - 1.99
# uva - 4.56
# beterraba - 7.5
# ovos - 10
# bacon - 12.99
# pão integral - 8.38
dict_precos = {'banana': 2.99, 'maça': 4.99, 'uva': 4.56, 'beterraba': 7.5, 'ovos': 10, 'bacon': 12.99, 'pão integral': 8.38}

# Defina uma função chamada 'comprar_carrinho' que receba uma lista de alimentos e retorne o valor total da compra. 
# Ignore termos ou não o item em estoque. Teste a função.
def comprar_carrinho(lst):
    valor_total = 0
    for i in lst:
        valor_total += dict_precos.get(i, 0)
        
    return valor_total

lista_compras = ['banana', 'uva', 'melão', 'ovos', 'pão integral', 'maça']
print(format('sem considerar que há o item em estoque','*^82'))
print(f'Total de compras: {comprar_carrinho(lista_compras)}', end='\n\n')


# Altere a função 'comprar_carrinho' para que só sejam computados os valores dos items em estoque. 
# Caso o item da lista esteja em estoque, após somar seu valor ao total da compra diminua o estoque na quantidade necessária. 
# Retorne o valor total da compra e o dicionário de estoque atualizado.
def comprar_carrinho(lst):
    valor_total = 0
    for i in lst:
        if dict_estoque.get(i, 0) > 0:
            valor_total += dict_precos.get(i, 0)
            dict_estoque[i] -= 1
        
    return valor_total, dict_estoque

total_compras, estoque_atualizado = comprar_carrinho(lista_compras)

print(format('considerando apenas os itens em estoque','*^82'))
print(f'Total de compras: {total_compras}')
print(f'Estoque atualizado: {estoque_atualizado}')
```
